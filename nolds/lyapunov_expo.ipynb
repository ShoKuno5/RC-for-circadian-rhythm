```python
import pandas as pd
import nolds
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# リアプノフ指数を格納するための辞書
lyap_exp_values = {'opt_shift_hour': [], 'x': [], 'y': [], 'P_shifted': []}

# opt_shift_hourが-12から12の場合
for opt_shift_hour in range(-12, 13):
    # ファイルパスの設定
    file_path = f'/home/kuno/my_project/VDP/VDP_analysis/generate_data/data/VDP_{opt_shift_hour}.csv'
    # データの読み込み
    vdp_data = pd.read_csv(file_path)[:10000]

    # 各列に対してリアプノフ指数を計算
    for column in ['x', 'y', 'P_shifted']:
        data = vdp_data[column].values
        lyap_exp = nolds.lyap_r(data, min_tsep=10, tau=2, emb_dim=3, min_neighbors=20, trajectory_len=20)
        lyap_exp_values[column].append(lyap_exp)
    
    lyap_exp_values['opt_shift_hour'].append(opt_shift_hour)

# opt_shift_hourがrandomの場合
file_path = '/home/kuno/my_project/VDP/VDP_analysis/0random/generate_data/data/VDP_random.csv'
vdp_data = pd.read_csv(file_path)[:1000]

# 各列に対してリアプノフ指数を計算
for column in ['x', 'y', 'P_shifted']:
    data = vdp_data[column].values
    lyap_exp = nolds.lyap_r(data, min_tsep=10, tau=2, emb_dim=3, min_neighbors=20, trajectory_len=20)
    lyap_exp_values[column].append(lyap_exp)

lyap_exp_values['opt_shift_hour'].append('  random')

# グラフの作成
plt.figure(figsize=(12, 6))
plt.plot(lyap_exp_values['opt_shift_hour'], lyap_exp_values['x'], label='x')
plt.plot(lyap_exp_values['opt_shift_hour'], lyap_exp_values['y'], label='y')
plt.plot(lyap_exp_values['opt_shift_hour'], lyap_exp_values['P_shifted'], label='P_shifted')

plt.xlabel('opt_shift_hour')
plt.ylabel('Lyapunov Exponent')
plt.title('Lyapunov Exponent for Different opt_shift_hours')
plt.legend()

# PDF形式で保存
pdf_file_path = '/home/kuno/my_project/VDP/VDP_analysis/nolds/plot/lyap_graph.pdf'  # 保存先のパスを設定
with PdfPages(pdf_file_path) as pdf:
    pdf.savefig()  # 現在の図をPDFに保存
    plt.close()  # 図を閉じる
```
